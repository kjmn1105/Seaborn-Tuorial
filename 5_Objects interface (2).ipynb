# Mark object 의 속성
Properties of Mark objects

## 1. 좌표 속성
Coordinate properties

### x, y, xmin, xmax, ymin, ymax
<br>
좌표 속성은 플롯에서 마크가 그려지는 위치를 결정합니다. 
일반적으로 `x` 좌표는 수평, `y` 좌표는 수직에 위치합니다. 
일부 mark는 하나 또는 두 변수에 대해 스팬(span; ex. `min`, `max`) 매개변수화를 허용합니다. 
그 외의 mark는 `x`와 `y`를 허용하지만 기준선(`baseline`) 매개변수를 사용하여 스팬을 표시할 수도 있습니다. 
해당 layer의 방향(`orient`) 매개변수에 따라 이 작동 방식이 결정됩니다.   
<br></br>
변수(variable)에 수치형 데이터가 포함되지 않은 경우, scale은 데이터가 화면에 그려질 수 있는 규칙(convention)을 적용합니다. 
예를 들어, 명목척도(`Nominal` scale)는 각각의 범주(category)에 정수 인덱스를 할당하고, 시간척도(`Temporal` scale)는 "기준시간(reference epoch)"을 기준으로부터의 일수를 나타냅니다.  
<br></br>  


<br></br>
연속형척도(`Continuous` scale)는 데이터 값과 공간위치(spatial position) 간에 비선형 변환을 적용할 수도 있습니다.
<br></br>  





## 2. 색상 속성
Color properties

### 색, 채우기색, 선색
color, fillcolor, edgecolor  
<br>

모든 mark는 `color`를 지정할 수 있고, 많은 경우 "선(edge) 색"과 "채우기(fill) 색"을 구분해서 지정합니다. 
단순히 '색(color)"를 사용하면 이 두 가지가 알아서 지정되기도 하지만, 구체적인 속성을 사용하면 더 많은 제어가 가능합니다.  
<br>  


<br>

색상 속성(color property)가 맵핑된 경우, 기본(default) palette는 scale의 타입에 따라 달라집니다. 
명목(Nominal)scale는 정렬되지 않는 색상(hue)를 사용하는 반면, 연속(temporal 포함)scale는 순차적인 그라데이션(sequential gradient)를 사용합니다.  
<br>


<br>  

(연속 scale의 기본값은 향후 버전에서는 식별성 개선을 위해 변경될 수 있습니다.)  
<br>  

색상 scale은 `'viridis'`, `'rocket'`, `'deep'` 처럼 palette의 이름으로 매개변수화(parameterized) 됩니다.
palette의 이름은 그래디어트(ex. `'dark:blue'`), 큐브헬릭스 시스템(cubehelix; ex.`'ch:start=.2, rot=-.4'`)처럼 parameter를 포함할 수도 있습니다.
더 자세한 내용은 [color palette tutorial](https://seaborn.pydata.org/tutorial/color_palettes.html) 을 참고하세요.  

<br>  
  
연속(continuous) scale은 색상의 tuple로 매개변수화될 수도 있습니다. 
명목(nominal) scale을 사용하는 경우 palette의 이름(필요한 경우 개별적으로 샘플링된), 개별 color 값들의 list, 혹은 data값을 색상에 맵핑하는 dictionary를 사용할 수 있습니다.  
<br>  

개별 색상은 [다양한 형식](https://matplotlib.org/stable/tutorials/colors/colors.html)으로 지정할 수 있습니다 : 
- 색상주기(color cycle)에 대한 색인참조 : ex. `'C0'`, 
- 한 글자 약어 : ex. `'b'`, 
- 회색조(grayscale)값 : ex. `'.4'`
- RGB hex 코드 : ex. `'#4c72b0' `
- X11 색상 이름 : ex. `'seagreen'`
- XKCD color survey 이름 : ex. `'purpleish'`    
<br><br>    
    



### 투명도, 채우기 투명도, 선 투명도
alpha, fillalpha, edgealpha

`alpha` property는 mark의 투명도(opacity)를 결정하는 속성입니다. 
표현해야 하는 데이터가 너무 많아서 mark가 겹치면(overplotting), alpha 값을 낮춰서 밀도를 표현하는 것이 좋습니다.
<br>  
  
<br>
alpha는 mark가 겹치지 않을 때에도 중요도를 표현하거나, `color`와 함께 사용해서 2개의 변수를 표현하는데 사용할 수 있습니다.
또, alpha 값이 낮으면 채도가 낮아져서 bar 처럼 큰 mark를 더 보기 좋게 채워줍니다.
<br>  
`color`와 마찬가지로, mark의 선투명도(`edgealph`)와 채우기투명도(`fillalpha`) 속성을 별도로 정의해서 제어하는 경우도 있습니다.
