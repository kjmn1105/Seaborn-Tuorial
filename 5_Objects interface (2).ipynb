# Mark object 의 속성
Properties of Mark objects

## 1. 좌표 속성
Coordinate properties

### x, y, xmin, xmax, ymin, ymax
<br>
좌표 속성은 플롯에서 마크가 그려지는 위치를 결정합니다. 
일반적으로 `x` 좌표는 수평, `y` 좌표는 수직에 위치합니다. 
일부 mark는 하나 또는 두 변수에 대해 스팬(span; ex. `min`, `max`) 매개변수화를 허용합니다. 
그 외의 mark는 `x`와 `y`를 허용하지만 기준선(`baseline`) 매개변수를 사용하여 스팬을 표시할 수도 있습니다. 
해당 layer의 방향(`orient`) 매개변수에 따라 이 작동 방식이 결정됩니다.   
<br></br>
변수(variable)에 수치형 데이터가 포함되지 않은 경우, scale은 데이터가 화면에 그려질 수 있는 규칙(convention)을 적용합니다. 
예를 들어, 명목척도(`Nominal` scale)는 각각의 범주(category)에 정수 인덱스를 할당하고, 시간척도(`Temporal` scale)는 "기준시간(reference epoch)"을 기준으로부터의 일수를 나타냅니다.  
<br></br>  


<br></br>
연속형척도(`Continuous` scale)는 데이터 값과 공간위치(spatial position) 간에 비선형 변환을 적용할 수도 있습니다.
<br></br>  





## 2. 색상 속성
Color properties

### 색, 채우기색, 선색
color, fillcolor, edgecolor  
<br>
모든 mark는 `color`를 지정할 수 있고, 많은 경우 "선(edge) 색"과 "채우기(fill) 색"을 구분해서 지정합니다. 
단순히 '색(color)"를 사용하면 이 두 가지가 알아서 지정되기도 하지만, 구체적인 속성을 사용하면 더 많은 제어가 가능합니다.
